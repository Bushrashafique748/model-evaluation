```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = pd.read_csv('/content/tested.csv')

# Display the first few rows of the dataframe
print(df.head())
```

```python
# Understanding the structure of the data
print(df.info())

# Statistical summary of the data
print(df.describe())

# Examine the features in the dataset
print(df.columns)
```

```python
# Handling missing values
print(df.isnull().sum())

# Fill missing values in 'Age' with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop rows with missing values in 'Embarked'
df.dropna(subset=['Embarked'], inplace=True)

# Drop the 'Cabin' column due to a high number of missing values
df.drop(columns=['Cabin'], inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Verify no missing values remain
print(df.isnull().sum())
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Visualization 1: Distribution of Ages
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Purpose: To understand the age distribution of passengers
# Insight: The majority of passengers are between 20 and 40 years old.

# Visualization 2: Survival rate by Sex
plt.figure(figsize=(6, 4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()

# Purpose: To compare survival rates between male and female passengers
# Insight: Females had a significantly higher survival rate than males.

# Visualization 3: Survival rate by Pclass
plt.figure(figsize=(6, 4))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Purpose: To analyze survival rates across different passenger classes
# Insight: Passengers in first class had the highest survival rate, followed by second and third classes.

# Visualization 4: Age distribution by Pclass
plt.figure(figsize=(6, 4))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()

# Purpose: To understand the age distribution within each passenger class
# Insight: First class passengers tend to be older than those in second and third classes.
```

#Analysis and Insights


**Age Distribution:** Most passengers were aged between 20 and 40, indicating the primary age group of travelers.

**Survival Rate by Sex:** Females had a significantly higher survival rate than males, likely due to the "women and children first" protocol.

**Survival Rate by Passenger Class:** First class passengers had the highest survival rate, indicating socioeconomic status influenced survival chances.

**Age Distribution by Passenger Class:**First class passengers were generally older, while third class had more younger passengers and children.
